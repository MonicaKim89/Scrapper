```python
path = "C:\\Users\\user\\Documents\\Monica's\\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.implicitly_wait(5)

start_url = "https://www.tripadvisor.com/Search?q=medical%20tourism&searchSessionId=877DAD5D427AC111693CAD907C5181E71679364132653ssid&sid=26B75593323745DD8A8576EE9AA852311679364612762&blockRedirect=true&ssrc=f&isSingleSearch=true&searchNearby=true&geo=1"
driver.get(start_url)
```

```python
topic_name = []
forum_location= []
question  = []
links =[]
k = 0
```

```python
try:
    while True:
        time.sleep(5)
        driver.implicitly_wait(10)
        page = driver.page_source
        soup = BeautifulSoup(page, 'html.parser')
        # driver.implicitly_wait(10)
        ## 데이터추출
        topic_name_list = soup.find_all('div', class_ ="forum-topic-name")
        forum_location_list = soup.find_all('div', class_ ="parent-forum")
        question_list = soup.find_all('span', class_ ="forum-snippet-text")
        links_list = soup.find_all('div',class_='result forum-result-card')

        ## 데이터정리
        for i in links_list:
            link = str(i)
            link = link.split("this, '")[1]
            link = link.split("', {")[0]
            link = "https://www.tripadvisor.com"+link
            links.append(link)

        for i in topic_name_list:
            text = i.text
            topic_name.append(text)

        for i in forum_location_list:
            text = i.text
            forum_location.append(text)


        for i in question_list:
            text = i.text
            question.append(text)
        
        button = driver.find_element(By.XPATH,r'//*//*[@id="BODY_BLOCK_JQUERY_REFLOW"]/div[2]/div/div[2]/div/div/div/div/div[1]/div/div[1]/div/div[4]/div/div[2]/div/div/div/a[2]').click()

        k += 1
        print(34-k)
except:
    print('에러페이지: ', k+1)
    pass

dataset  = pd.DataFrame({'topic_name':topic_name, 'location':forum_location, 'question': question, 'link':links})
dataset.head()
```

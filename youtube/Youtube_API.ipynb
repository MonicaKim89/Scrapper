##### 날짜 지정하고,total 수 뽑고, statistics뽑기

```python
search_response_210301_210331 = \
youtube.search().list(part = 'snippet', q='코로나 백신', type='video',
                                        maxResults =50, order = 'viewCount', 
#                                         pageToken ="CJYBEAA",
                                        publishedAfter='2021-02-26T00:00:00Z',
                                       publishedBefore='2021-03-31T00:00:00Z',
                                       videoDuration='any',
                     ).execute()

search_response_210301_210331
```

```python
# 영상제목_df_search_response_210301_210331_50
titles=[]
total_publishTime=[]
total_channel_title = []
for i in search_response_210301_210331['items']:
    titles.append(i['snippet']['title'])
    total_publishTime.append(i['snippet']['publishTime'])
    total_channel_title.append(i['snippet']['channelTitle'])
    
titles = pd.DataFrame(titles)
total_publishTime = pd.DataFrame(total_publishTime)
total_channel_title = pd.DataFrame(total_channel_title)

df_search_response_210301_210331_50 = pd.concat([titles, total_publishTime, total_channel_title], axis=1)
df_search_response_210301_210331_50
```

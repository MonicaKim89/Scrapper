```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
import selenium.common.exceptions as sex

import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import lxml

import time
from time import sleep
import random
```

```python
path = "C:\\chromedriver.exe"
driver = webdriver.Chrome(path)
# driver.implicitly_waits(5)

start_url = 'https://www.modoodoc.com/hospitals/?search_query=%EC%B9%98%EA%B3%BC&page=1'
driver.get(start_url)
```

```python
root_url = 'https://www.modoodoc.com/hospitals/?search_query=%EC%B9%98%EA%B3%BC&page='
link_url = []

for i in range(1,1000):
    num = str(i)
    link = root_url+num
    link_url.append(link)
```

```python
page = driver.page_source
soup = BeautifulSoup(page, 'html.parser')
```

```python
name = []
star_score = []
address=[]
review = []
spec = []
link_dental = []
http = 'https://www.modoodoc.com'
k=0

for i in link_url:
    try : 
        driver.get(i)
        page = driver.page_source
        soup = BeautifulSoup(page, 'html.parser')


        names = soup.find_all('div', style="font-size: 1.313rem; font-weight:bold")
        for i in names:
            i = i.text
            i = i.strip()
            name.append(i)
            # print(i)

        stars = soup.find_all('div', class_="hospitals-review-score ml-1 mr-2")
        for i in stars:
            i = i.text
            i = i.strip()
            star_score.append(i)
            # print(i)

        addresses = soup.find_all('div', class_="my-1 hospitals-address")
        for i in addresses:
            i = i.text
            i = i.strip()
            address.append(i)
            # print(i)

        reviews = soup.find_all('div', class_="hospitals-review-count")
        for i in reviews:
            i = i.text
            i = i.strip()
            i = i.split('리뷰 ')[-1]
            review.append(i)
            # print(i)

        specs = soup.find_all('div', class_="row px-3 mb-2 pl-4")
        for i in specs:
            i = i.text
            i = i.strip()
            i = i.split('리뷰 ')[-1]
            spec.append(i)

        links= soup.find_all('div', class_ = "doctor-total-box border-bottom")

        for i in links:
        #     link_list.append(i)
            links_=(i.find('a')['href'])
            link_dental.append(http+links_)

        x = random.randint(0,1)
        sleep(x)
            
        print('sleep for ', x)

        k += 1
        print(k,'번쨰 완료')

    except:
        print(i)
        pass

print(len(name))
print(len(star_score))
print(len(address))
print(len(review))
print(len(spec))
print(len(link_dental))



import pickle
with open ('name.pkl','wb') as f:
    pickle.dump(name, f)

with open ('star_score.pkl','wb') as f:
    pickle.dump(star_score, f)

with open ('address.pkl','wb') as f:
    pickle.dump(address, f)

with open ('review.pkl','wb') as f:
    pickle.dump(review, f)

with open ('spec.pkl','wb') as f:
    pickle.dump(spec, f)

with open ('link_dental.pkl','wb') as f:
    pickle.dump(link_dental, f)
```

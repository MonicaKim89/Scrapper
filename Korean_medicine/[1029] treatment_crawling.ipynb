```python
treatments = []
hospitals = []
adr = []

z =0

for num, k in enumerate (url_link[1053:]):
    try:
        driver.get(k)
        page = driver.page_source
        soup = BeautifulSoup(page, 'html.parser')

        treatment = soup.find_all('div', class_ ="review-tc-filter container mt-4 mb-2 p-3")
        for i in treatment:
            i = i.text
            i = i.strip()
            i = i.replace('・', '')
            i = i.lstrip()
            i = i.rstrip()
            i = i.replace(' ','')
            i = i.replace('\n', '')
            treatments.append(i)

        hospital = soup.find('h1', class_ ="my-1 hospital-doctor-name-box d-flex align-items-center" ).text
        hospital = hospital.lstrip()
        hospital = hospital.lstrip()
        hospital = hospital.replace('\n', '')
        hospitals.append(hospital)

        adress = soup.find('div', class_ ="col-20 text-left d-flex align-items-center justify-content-between pt-2").text
        adress = adress.lstrip()
        adress = adress.split(' ')[0]
        adr.append(adress)

        z+=1

        x = random.randint(0,2)
    #     sleep(x)
        print(i, hospital, adress)
        print ('----------------------{}-----------------'.format(len(url_link)-1053- z))
        
    except:
        print('조회안됨', num)
        driver.get(start_url)
        pass
```

```python
data1 = pd.DataFrame({'hospital':hospitals[:-1], 'adress':adr[:-1], 'treatments':treatments})
data1.to_csv('frequency_treatment1s.csv')
```
